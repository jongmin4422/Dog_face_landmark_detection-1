```python
import tensorflow as tf
from tensorflow import keras
import data_util
import matplotlib.pyplot as plt
```

```python
train_ds, test_ds = data_util.get_autoencoder_dataset()
```

```python
sample = None
for i in train_ds.take(1):
    sample = i
sample.shape
```

```python
drop = keras.layers.Dropout(0.5)
```

```python
drop_sample = drop(sample, training=True)
```

```python
for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(sample[i].numpy())
        plt.axis('off')
```

```python
for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(drop_sample[i].numpy())
        plt.axis('off')
```
