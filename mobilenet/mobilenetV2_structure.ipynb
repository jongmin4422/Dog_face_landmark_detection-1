```python
from tensorflow import keras
from tensorflow.keras.utils import plot_model
```

```python
mobilenet = keras.applications.MobileNetV2(input_shape=(256,256,3), alpha=1.0, include_top=False)
```

```python
mobilenet.summary()
```

```python
plot_model(mobilenet, to_file='img/MobileNetV2.png', show_shapes=True)
```
